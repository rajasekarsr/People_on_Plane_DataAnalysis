##### Let us consider both these categorical variables,

We are not interested to know if there is a relationship between the person gender and their tendency to accept a baby on a plane. In order to do so, we would have to use the Chi-Square test. But First, let's state our null and alternate hypothesis. 

Ho : There is no staistically significant relationship between these two variables.
H1 : There is a staistically significant relationship between these two variables.

In order to perfrom this test, we need to construct the contigency table for these two variables.

```python
contigency_table_flying = pd.crosstab(
filled_flying['Gender'],
filled_flying['In general, is itrude to bring a baby on a plane?'],
margins = True)
contigency_table_flying
```

Now we can see the counts of the people who are very rude towards bringing baby on a flight. Each row in the above table will represents the count. The intersection of Male and No, not at all rude represent that there are 249 Male who think bring a baby on plane is not rude. 

```python
#assign the gender counts to variables
male_count = contigency_table_flying.iloc[1][0:3].values
female_count = contigency_table_flying.iloc[0][0:3].values
unknown_count = contigency_table_flying.iloc[2][0:3].values

print (male_count, female_count, unknown_count)
```

```python
#Plots the bar chart
fig = plt.figure(figsize=(10, 5))
sns.set(font_scale=1.8)
categories = ["No, not at all rude","Yes, Somewhat rude","Yes, very rude"]
p1 = plt.bar(categories, male_count, 0.55, color='#d62728')
p2 = plt.bar(categories, female_count, 0.55, bottom= male_count)
plt.legend((p2[0], p1[0]), ('Female', 'Male'))
plt.xlabel('Having Baby on plane')
plt.ylabel('Count')
plt.show()
```

```python
f_obs = np.array ([contigency_table_flying.iloc[1][0:3].values,
                  contigency_table_flying.iloc[0][0:3].values])
f_obs
```

```python
from scipy import stats
chisq = stats.chi2_contingency(f_obs)[0:3]
print 'Chi-square value = ', chisq[0],'\n P-value = ', '{:.6f}'.format (round (float (chisq[1]), 5))
```

#### As we can see, the p-value = 0.000030, (i.e) (p < 0.05) we can reject the null hypothesis. 

There is definietly some sort of relationship between sex and being rude towards having baby on the plane.
